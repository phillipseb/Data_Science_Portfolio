## Modeling

Since there are 45 columns (46 without the target variable) in the dataset, I am going to use a feature reduction technique to reduce the number of features that are included in the classification models.

```python
merged_metro.dtypes
```

For modeling purposes, all variables have to be numeric, so I am going to create a separate dataset for modeling purposes that has 'Cbsa_Title' removed since it is the location descriptor for the 'Cbsa_Code' variable.

Our target variable for modeling is Hotness Rank ... to try and predict the hotness of a given metro area based on the other factors in the dataset. Given that rank is actually a categorical variable (hierarchical), we will be using categorical modeling approaches (classification) to solve the problem.

```python
model_metro = merged_metro.drop(['Cbsa_Title'],axis=1)
```

```python
print(model_metro.shape)
```

The multi-class distribution of the target variable seems relatively even! There is not a startling contrast that I am concerned about and need to handle before modeling.

```python
#matrices of features
X = model_metro.drop(labels=['Hotness_Rank'],axis=1)
y = model_metro['Hotness_Rank']
col=X.columns
```

```python
print(col)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#feature reduction
# Building the model --> extra tree classifier for feature reduction
#finds Gini Importance as selection criteria
extra_tree_forest = ExtraTreesClassifier(n_estimators = 5,
                                        criterion ='entropy', max_features = 2)
  
# Training the model
extra_tree_forest.fit(X, y)
  
# Computing the importance of each feature
feature_importance = extra_tree_forest.feature_importances_
  
# Normalizing the individual importances
feature_importance_normalized = np.std([tree.feature_importances_ for tree in 
                                        extra_tree_forest.estimators_],
                                        axis = 0)
```

```python
# Plotting a Bar Graph to compare the models
plt.bar(X.columns, feature_importance_normalized)
plt.xlabel('Feature Labels')
plt.ylabel('Feature Importances')
plt.title('Comparison of different Feature Importances')
plt.show()
```

I moved the features and their corresponding feature importance values to a separate document, and found that there are five values with an importance measure greater than 0.008 which I am using as my cutoff.

These are the following top five features:
1. Median_Dom_Vs_Us
2. Cbsa_code
3. 2021_estimate
4. Median_Listing_Price_Mm
5. Total_Listing_Count_Yy

I am now going to reduce the model to only use these features and continue with modeling.

```python
model_metro = model_metro.drop(model_metro.columns.difference(['Hotness_Rank','Median_Dom_Vs_Us',
'Cbsa_code',
'2021_estimate',
'Median_Listing_Price_Mm',
'Total_Listing_Count_Yy']),1)
```

```python
print(model_metro.shape)
```

```python
#split into features vs. target once more
#matrices of features
X_reduced = model_metro.drop(labels=['Hotness_Rank'],axis=1)
y_reduced = model_metro['Hotness_Rank']
col=X_reduced.columns
```

```python
#train, test, split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_reduced,y_reduced,test_size=0.2,random_state= 10)
```

```python
#scale data to [-1,1]

from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler(feature_range=(-1,1)).fit(X_train)
X_train = scaling.transform(X_train)
X_test = scaling.transform(X_test)
```

```python
print("X_train shape: ",X_train.shape)
print("y train shape: ",y_train.shape)
print("X test shape: ", X_test.shape)
print("y test shape: ",y_test.shape)
```

```python
#looking at class distribution
#histogram
plt.hist(y_train)
plt.title('Histogram of Hotness Rank')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# training a linear SVM classifier
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)
```

### Support Vector Machine (SVM)

```python
from sklearn.metrics import precision_recall_fscore_support

# model accuracy for X_test 
svm_accuracy = svm_model_linear.score(X_test, y_test)
print(svm_accuracy)
 
# creating a confusion matrix
svm_cm = confusion_matrix(y_test, svm_predictions)
print(svm_cm)

#getting precision and recall
svm_metrics = precision_recall_fscore_support(y_test, svm_predictions, average='macro',labels=np.unique(svm_predictions),zero_division=1)
print(svm_metrics)
```

### K-Nearest Neighbor

```python
# training a KNN classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)
```

```python
# accuracy on X_test
knn_accuracy = knn.score(X_test, y_test)
print(knn_accuracy)
 
# creating a confusion matrix
knn_predictions = knn.predict(X_test)
knn_cm = confusion_matrix(y_test, knn_predictions)

#getting precision and recall
knn_metrics = precision_recall_fscore_support(y_test, knn_predictions, average='macro',labels=np.unique(knn_predictions),zero_division=1)
print(knn_metrics)
```

### Naive Bayes Classifier

```python
# training a Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)
```

```python
# accuracy on X_test
gnb_accuracy = gnb.score(X_test, y_test)
print(gnb_accuracy)
 
# creating a confusion matrix
gnb_cm = confusion_matrix(y_test, gnb_predictions)

#getting precision and recall
gnb_metrics = precision_recall_fscore_support(y_test, gnb_predictions, average='macro',labels=np.unique(gnb_predictions),zero_division=1)
print(gnb_metrics)
```

### Multinomial Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RepeatedStratifiedKFold

# define the multinomial logistic regression model
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
# define the model evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate the model and collect the scores
n_scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
# report the model performance
print('Mean Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

```python
# fit the model on the whole dataset
model.fit(X_train, y_train)
#get predictions
yhat = model.predict(X_test)
# summarize the predicted class
print('Predicted Class: %d' % yhat[0])
```
